```python
#lets eliminate null values from the dataset

dc.drop(columns = 'Cabin', axis = 1, inplace = True)
dc.columns
```

**FEATURE ENCODING - LABEL ENCODING**

```python
# lets check the coorelation of all the columns

rev_dics = {"YES":1,"NO":0}
dc["Survived"] = dc["Survived"].map(rev_dics)
```

```python
dc
```

```python
# lets make male = 1, female = 0

dicg = {"male":1,"female":0}
dc["Sex"] = dc["Sex"].map(dicg)
dc
```

```python
# lets convert embarked to numerics using ordinal encoding

dice = {"S":2,"C":1,"Q":0}
dc["Embarked"] = dc["Embarked"].map(dice)
dc
```

```python
# lets drop all the unnecessary columns which are strings

dc.drop(["PassengerId","Name","Ticket",],axis=1,inplace=True)
dc
```

**LOGISTIC REGRESSION**

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
x = dc.drop(columns="Survived",axis=1)
y = dc["Survived"]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train
```

```python
y_test
```

```python
lr = LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr_score = lr.score(x_test,y_test)
```

```python
accuracy_score(y_test,lr.predict(x_test))
```
